```python
import numpy as np
from keras.layers import Dense
from keras.models import Sequential,load_model
```

```python
dataset = np.loadtxt('data.csv',delimiter = ',')
data,test = dataset[:int(len(dataset)*95/100)],dataset [int (len (dataset)*95/100):]
X = data [:,0:8]
y = data[:,8]
```

Number of times pregnant
Plasma glucose concentration a 2 hours in an oral glucose tolerance test
Diastolic blood pressure (mm Hg)
Triceps skin fold thickness (mm)
2-Hour serum insulin (mu U/ml)
Body mass index (weight in kg/(height in m)^2)
Diabetes pedigree function
Age (years)

```python
# creazione modello keras
model = Sequential ()
model.add (Dense (16,activation = 'relu',input_dim = 8))
model.add (Dense (8,activation = 'relu'))
model.add (Dense (1,activation = 'sigmoid'))
model.compile (loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.fit (X,y,epochs = 175, batch_size = 32)
```

```python
Xt = test[:,0:8]
yt = test[:,8]
```

```python
def use_model (inputs):
    inputs_processati = np.array ([inputs])
    return model.predict (inputs_processati)[0][0]
```

```python
for i in range (len(Xt)):
    print (f'predetto: {use_model (Xt[i])} \t label: {yt[i]}')
```

```python
model.save ('model.h5')
```
